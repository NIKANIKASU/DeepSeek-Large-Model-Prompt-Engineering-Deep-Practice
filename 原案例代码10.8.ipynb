```python
import requests
import json

api_url = "https://api.deepseek.com/v1/chat/completions"  #文档里如果不是 v1，就改成实际路径
api_key = input("请输入您的api秘钥：")                           # 你的密钥

class TextGenerationClient:
    def __init__(self, api_url, api_key):
        self.api_url = api_url
        self.api_key = api_key

    def generate_text(self, payload):
        headers = {
            "Content-Type": "application/json",
            "Authorization": f"Bearer {self.api_key}",
        }
        # 也可以写成 requests.post(self.api_url, headers=headers, json=payload)
        response = requests.post(self.api_url, headers=headers, data=json.dumps(payload), timeout=60)

        if response.status_code == 200:
            data = response.json()
            # OpenAI 兼容格式：choices[0].message.content
            return data["choices"][0]["message"]["content"]
        else:
            # 报错时把服务器返回原文带出来，便于排查
            raise Exception(f"Error: {response.status_code} - {response.text}")

client = TextGenerationClient(api_url,api_key)
payload = {
    "model": "deepseek-chat",
    "messages": [
    {"role": "system", "content": "你是一个短片小说创作者，写一篇短篇小说，500字以内"},
    {"role": "user", "content": "3个人物，2个情节"}
    ],
    "max_tokens": 800,
    "top_p": 1,
    "temperature": 1,
}
generated_text = client.generate_text(payload)
print(generated_text)
content = generated_text #content就是指文章

payload = {
    "model": "deepseek-chat",
    "messages": [
        {"role": "system", "content": f"你是一个文本阅读器，读取文章的内容并按照用户的要求输出对于文章的回答，以下是文章{content}"},
        {"role": "user", "content": "根据文章回答，这个故事里的情感色彩是什么样的？"}
    ],
    "max_tokens": 200,
    "top_p": 1,
    "temperature": 0.3,
}
generated_text = client.generate_text(payload)
print(generated_text) #输出了上面的内容


payload = {
    "model": "deepseek-chat",
    "messages": [
    {"role": "system", "content": f"你是一个文本阅读器，读取文章的内容并按照用户的要求输出对于文章的回答，以下是文章{content}"},
    {"role": "user", "content": "根据文章回答：文章中有几个人物，并且分别列出来人物的详细信息"}
    ],
    "max_tokens": 200,
    "top_p": 1,
    "temperature": 1,
}
generated_text = client.generate_text(payload)
print(generated_text)        
```
